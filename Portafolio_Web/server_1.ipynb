```python
# ============================================================
# [RESTO DEL C√ìDIGO PYTHON DEL SERVIDOR]
# ============================================================

class MetGoHandler(http.server.SimpleHTTPRequestHandler):
    """Handler personalizado que inyecta navbar profesional"""
    
    def __init__(self, *args, **kwargs):
        super().__init__(*args, directory=DIRECTORY, **kwargs)
    
    def do_GET(self):
        """Procesa peticiones GET e inyecta navbar"""
        
        # Mapeo de rutas
        path_mapping = {
            '/': '/index.html',
            '/cv': '/cv.html',
            '/dashboard': '/dashboard_marca_personal.html',
            '/inicio': '/index.html'
        }
        
        # Normalizar path
        if self.path in path_mapping:
            self.path = path_mapping[self.path]
        
        # Verificar si es archivo HTML
        if self.path.endswith('.html'):
            try:
                file_path = DIRECTORY + self.path
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                
                # Inyectar navbar profesional
                modified_content = self.inject_navbar(content, self.path)
                
                # Enviar respuesta
                self.send_response(200)
                self.send_header('Content-type', 'text/html; charset=utf-8')
                self.send_header('Content-Length', len(modified_content.encode('utf-8')))
                self.end_headers()
                self.wfile.write(modified_content.encode('utf-8'))
                
                print(f"‚úÖ Servido: {self.path}")
                return
                
            except FileNotFoundError:
                self.send_error(404, f"Archivo no encontrado: {self.path}")
                return
        
        # Para otros archivos (CSS, JS, im√°genes)
        super().do_GET()
    
    def do_POST(self):
        """Procesa peticiones POST del formulario de contacto"""
        
        if self.path == '/enviar-consulta':
            try:
                # Leer los datos del formulario
                content_length = int(self.headers.get('Content-Length', 0))
                if content_length == 0:
                    self.send_error(400, "No se recibieron datos")
                    return
                
                post_data = self.rfile.read(content_length).decode('utf-8')
                
                print(f"DEBUG: Datos recibidos: {post_data[:200]}")  # Solo primeros 200 chars
                
                # Parsear los datos del formulario (application/x-www-form-urlencoded)
                form_data = parse_qs(post_data, keep_blank_values=True)
                
                # Extraer los campos de forma segura
                name = form_data.get('name', [''])[0] if form_data.get('name') else ''
                email = form_data.get('email', [''])[0] if form_data.get('email') else ''
                company = form_data.get('company', [''])[0] if form_data.get('company') else ''
                message = form_data.get('message', [''])[0] if form_data.get('message') else ''
                
                # Limpiar espacios en blanco
                name = name.strip() if name else ''
                email = email.strip() if email else ''
                company = company.strip() if company else ''
                message = message.strip() if message else ''
                
                print(f"DEBUG: Campos parseados - name: '{name}', email: '{email}', company: '{company}'")
                
                # Guardar en base de datos
                consulta_id = save_consulta(name, email, company, message)
                
                # Imprimir la consulta en consola
                print(f"\n{'='*70}")
                print(f"üìß NUEVA CONSULTA RECIBIDA (ID: {consulta_id})")
                print(f"{'='*70}")
                print(f"Nombre: {name}")
                print(f"Email: {email}")
                print(f"Empresa: {company}")
                print(f"Mensaje: {message}")
                print(f"Fecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
                print(f"{'='*70}\n")
                
                # Crear p√°gina de confirmaci√≥n
                confirmation_html = self.create_confirmation_page(name, email)
                
                # Enviar respuesta de confirmaci√≥n
                self.send_response(200)
                self.send_header('Content-type', 'text/html; charset=utf-8')
                self.send_header('Content-Length', len(confirmation_html.encode('utf-8')))
                self.end_headers()
                self.wfile.write(confirmation_html.encode('utf-8'))
                
                return
                
            except Exception as e:
                print(f"‚ùå Error al procesar formulario: {e}")
                self.send_error(500, f"Error al procesar el formulario: {str(e)}")
                return
        
        # Para otras rutas POST
        self.send_error(404, f"Ruta no encontrada: {self.path}")
    
    def create_confirmation_page(self, name, email):
        """Crea la p√°gina de confirmaci√≥n despu√©s de enviar el formulario"""
        
        html = f"""<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="utf-8">
    <title>Consulta Enviada - MetGo3d</title>
    <meta name="viewport" content="width=device-width, initial-scale=1">
    <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700&family=Playfair+Display:wght@400;700&display=swap" rel="stylesheet">
    <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.3/dist/css/bootstrap.min.css">
    <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap-icons@1.11.3/font/bootstrap-icons.css">
    
    <style>
        :root {{
            --metgo-dark: #003366;
            --metgo-accent: #FF4500;
            --metgo-light: #f8fafc;
        }}
        
        body {{
            font-family: 'Inter', system-ui, sans-serif;
            color: var(--metgo-dark);
            background-color: var(--metgo-light);
            display: flex;
            align-items: center;
            justify-content: center;
            min-height: 100vh;
            margin: 0;
        }}
        
        .confirmation-card {{
            background: white;
            padding: 3rem;
            border-radius: 8px;
            box-shadow: 0 10px 30px rgba(0,0,0,0.1);
            max-width: 600px;
            text-align: center;
            border-left: 4px solid var(--metgo-accent);
        }}
        
        .success-icon {{
            font-size: 4rem;
            color: #28a745;
            margin-bottom: 1rem;
        }}
        
        h1 {{
            color: var(--metgo-dark);
            margin-bottom: 1rem;
            font-family: 'Playfair Display', serif;
        }}
        
        .btn-primary {{
            background-color: var(--metgo-accent);
            border: 0;
            color: white;
            padding: 12px 35px;
            border-radius: 4px;
            font-weight: 600;
            transition: 0.3s;
            text-decoration: none;
            display: inline-block;
        }}
        
        .btn-primary:hover {{
            background-color: #e63e00;
            transform: translateY(-1px);
        }}
    </style>
</head>
<body>
    <div class="confirmation-card">
        <div class="success-icon">
            <i class="bi bi-check-circle-fill"></i>
        </div>
        <h1>¬°Consulta Enviada!</h1>
        <p class="lead">Gracias <strong>{name}</strong>, tu consulta ha sido recibida correctamente.</p>
        <p>Nos pondremos en contacto contigo a trav√©s de <strong>{email}</strong> en las pr√≥ximas 24-48 horas h√°biles.</p>
        <div class="mt-4">
            <a href="/index.html" class="btn-primary">Volver al Inicio</a>
            <a href="/formulario_contacto.html" class="btn-primary ms-3" style="background-color: var(--metgo-dark);">Enviar Otra Consulta</a>
        </div>
    </div>
</body>
    </html>"""
        
        return html
    
    def inject_navbar(self, html_content, current_path):
        """Inyecta navbar profesional en todas las p√°ginas"""
        
        # Determinar p√°gina activa
        active_index = 'active' if 'index' in current_path else ''
        active_cv = 'active' if 'cv' in current_path else ''
        active_dashboard = 'active' if 'dashboard' in current_path else ''
        
        navbar_html = f"""
<nav class="professional-nav">
    <div class="nav-container">
        <div class="nav-content">
            <a href="/index.html" class="nav-brand">Miguel Lucero</a>
            <ul class="nav-menu">
                <li><a href="/index.html" class="nav-link {active_index}">Inicio</a></li>
                <li><a href="/cv.html" class="nav-link {active_cv}">Curr√≠culum</a></li>
                <li><a href="/dashboard_marca_personal.html" class="nav-link {active_dashboard}">Dashboard</a></li>
            </ul>
        </div>
    </div>
</nav>
"""
        
        # Insertar navbar despu√©s del <body>
        if '<!-- Navbar inyectado autom√°ticamente por el servidor -->' in html_content:
            html_content = html_content.replace(
                '<!-- Navbar inyectado autom√°ticamente por el servidor -->',
                navbar_html
            )
        elif '<body' in html_content:
            html_content = re.sub(
                r'(<body[^>]*>)',
                r'\1\n' + navbar_html,
                html_content,
                count=1
            )
        
        return html_content
    
    def log_message(self, format, *args):
        """Personalizar mensajes de log"""
        print(f"üåê [{self.log_date_time_string()}] {format % args}")
```

```python
# ============================================================
# FUNCIONES AUXILIARES
# ============================================================
def create_html_files():
    """Crea los archivos HTML si no existen"""
    
    files = {
        'index.html': INDEX_HTML,
        'cv.html': CV_HTML,
        'formulario_contacto.html': FORMULARIO_CONTACTO_HTML,
        'dashboard_marca_personal.html': DASHBOARD_HTML,         
    }
    
    for filename, content in files.items():
        filepath = Path(DIRECTORY) / filename
        if not filepath.exists():
            with open(filepath, 'w', encoding='utf-8') as f:
                f.write(content)
            print(f"‚úÖ Creado: {filename}")
        else:
            print(f"‚ÑπÔ∏è  Ya existe: {filename}")


def create_assets_structure():
    """Crea la estructura de carpetas y archivos CSS/JS"""
    
    assets_dir = Path(DIRECTORY) / 'assets'
    css_dir = assets_dir / 'css'
    js_dir = assets_dir / 'js'
    
    # Crear directorios
    css_dir.mkdir(parents=True, exist_ok=True)
    js_dir.mkdir(parents=True, exist_ok=True)
    
    # Crear styles.css con dise√±o profesional
    styles_path = css_dir / 'styles.css'
    if not styles_path.exists():
        with open(styles_path, 'w', encoding='utf-8') as f:
            f.write(PROFESSIONAL_CSS)
        print(f"‚úÖ Creado: assets/css/styles.css")
    
    # Crear dashboard.js actualizado
    dashboard_js = js_dir / 'dashboard.js'
    if not dashboard_js.exists():
        dashboard_content = """
// Dashboard Profesional - Gr√°ficos con Plotly
document.addEventListener('DOMContentLoaded', function() {
    
    // Configuraci√≥n com√∫n profesional
    const commonLayout = {
        paper_bgcolor: 'rgba(0,0,0,0)',
        plot_bgcolor: 'rgba(247, 250, 252, 0.5)',
        font: {
            family: 'Inter, -apple-system, sans-serif',
            size: 12,
            color: '#4a5568'
        },
        margin: { t: 50, r: 30, b: 50, l: 50 }
    };
    
    const config = {
        responsive: true,
        displayModeBar: false
    };
    
    // Paleta de colores profesional
    const colors = {
        primary: '#1a365d',
        secondary: '#2c5282',
        accent: '#2b6cb0',
        success: '#38a169',
        warning: '#d69e2e',
        danger: '#e53e3e'
    };
    
    // 1. Distribuci√≥n de Contenidos por Mes
    if (document.getElementById('timeline-chart')) {
        const timelineData = [{
            x: ['Mes 1', 'Mes 2', 'Mes 3', 'Mes 4'],
            y: [2, 2, 2, 2],
            type: 'bar',
            name: 'Publicaciones',
            marker: {
                color: [colors.primary, colors.secondary, colors.accent, colors.primary],
                line: { color: '#fff', width: 2 }
            },
            text: ['2 casos', '2 casos', '2 casos', '2 casos'],
            textposition: 'outside',
            hovertemplate: '<b>%{x}</b><br>Publicaciones: %{y}<br>%{text}<extra></extra>'
        }];
        
        const timelineLayout = {
            ...commonLayout,
            title: {
                text: 'Distribuci√≥n Mensual de Publicaciones',
                font: { size: 16, color: colors.primary }
            },
            yaxis: {
                title: 'N√∫mero de Casos',
                gridcolor: '#e2e8f0',
                zeroline: false
            },
            xaxis: {
                title: '',
                tickfont: { size: 12 }
            }
        };
        
        Plotly.newPlot('timeline-chart', timelineData, timelineLayout, config);
    }
    
    // 2. Objetivos de Crecimiento
    if (document.getElementById('metrics-chart')) {
        const metricsData = [{
            x: ['Engagement', 'Crecimiento Red', 'Casos Totales'],
            y: [20, 30, 8],
            type: 'bar',
            marker: {
                color: [colors.success, colors.accent, colors.secondary]
            },
            text: ['20%', '+30%', '8 casos'],
            textposition: 'outside',
            hovertemplate: '<b>%{x}</b><br>Meta: %{text}<extra></extra>'
        }];
        
        const metricsLayout = {
            ...commonLayout,
            title: {
                text: 'Objetivos del Proyecto',
                font: { size: 16, color: colors.primary }
            },
            yaxis: {
                title: 'Valor Objetivo',
                gridcolor: '#e2e8f0',
                zeroline: false
            },
            xaxis: { title: '' }
        };
        
        Plotly.newPlot('metrics-chart', metricsData, metricsLayout, config);
    }
    
    // 3. Mix de Contenidos
    if (document.getElementById('content-chart')) {
        const contentData = [{
            values: [8, 4, 2, 3],
            labels: ['Casos de Estudio', 'Art√≠culos T√©cnicos', 'Videos', 'Infograf√≠as'],
            type: 'pie',
            marker: {
                colors: [colors.primary, colors.secondary, colors.accent, colors.success],
                line: { color: '#fff', width: 2 }
            },
            textfont: { size: 12, color: '#fff' },
            hovertemplate: '<b>%{label}</b><br>Cantidad: %{value}<br>%{percent}<extra></extra>'
        }];
        
        const contentLayout = {
            ...commonLayout,
            title: {
                text: 'Distribuci√≥n de Tipos de Contenido',
                font: { size: 16, color: colors.primary }
            },
            showlegend: true,
            legend: {
                orientation: 'h',
                yanchor: 'bottom',
                y: -0.2,
                xanchor: 'center',
                x: 0.5
            }
        };
        
        Plotly.newPlot('content-chart', contentData, contentLayout, config);
    }
    
    // 4. An√°lisis de Riesgos
    if (document.getElementById('risk-chart')) {
        const riskData = [{
            x: ['Bajo', 'Medio', 'Alto'],
            y: ['Tiempo', 'Recursos', 'Engagement'],
            z: [[1, 2, 1], [2, 3, 2], [1, 2, 3]],
            type: 'heatmap',
            colorscale: [
                [0, colors.success],
                [0.5, colors.warning],
                [1, colors.danger]
            ],
            showscale: true,
            hovertemplate: '<b>%{y}</b><br>Probabilidad: %{x}<br>Nivel: %{z}<extra></extra>',
            colorbar: {
                title: {
                    text: 'Nivel de<br>Riesgo',
                    side: 'right'
                },
                tickvals: [1, 2, 3],
                ticktext: ['Bajo', 'Medio', 'Alto'],
                thickness: 15
            }
        }];
        
        const riskLayout = {
            ...commonLayout,
            title: {
                text: 'Matriz de Evaluaci√≥n de Riesgos',
                font: { size: 16, color: colors.primary }
            },
            xaxis: {
                title: 'Probabilidad',
                side: 'bottom',
                tickfont: { size: 12 }
            },
            yaxis: {
                title: 'Factor de Riesgo',
                tickfont: { size: 12 }
            }
        };
        
        Plotly.newPlot('risk-chart', riskData, riskLayout, config);
    }
    
    // Hacer los gr√°ficos responsivos
    window.addEventListener('resize', function() {
        const charts = ['timeline-chart', 'metrics-chart', 'content-chart', 'risk-chart'];
        charts.forEach(chartId => {
            const chartDiv = document.getElementById(chartId);
            if (chartDiv && chartDiv.data) {
                Plotly.Plots.resize(chartDiv);
            }
        });
    });
});

// Manejo del formulario de contacto profesional
document.addEventListener('DOMContentLoaded', function() {
    const contactForm = document.querySelector('.contact-form');
    if (contactForm) {
        contactForm.addEventListener('submit', function(e) {
            e.preventDefault();
            
            const name = document.getElementById('name').value;
            const email = document.getElementById('email').value;
            const company = document.getElementById('company').value;
            
            // Mensaje profesional
            alert(`Gracias por su inter√©s, ${name}. Su consulta ha sido recibida y ser√° respondida en las pr√≥ximas 24-48 horas h√°biles a ${email}.`);
            
            // Limpiar formulario
            this.reset();
        });
    }
});
"""
        with open(dashboard_js, 'w', encoding='utf-8') as f:
            f.write(dashboard_content)
        print(f"‚úÖ Creado: assets/js/dashboard.js")
    
    # Crear scroll.js mejorado
    scroll_js = js_dir / 'scroll.js'
    if not scroll_js.exists():
        scroll_content = """
// Funcionalidad de scroll profesional

// Bot√≥n Scroll to Top
window.addEventListener('scroll', function() {
    const scrollBtn = document.getElementById('btnScrollTop');
    if (scrollBtn) {
        if (window.pageYOffset > 300) {
            scrollBtn.style.display = 'flex';
        } else {
            scrollBtn.style.display = 'none';
        }
    }
});

// Smooth scroll para enlaces internos
document.querySelectorAll('a[href^="#"]').forEach(anchor => {
    anchor.addEventListener('click', function (e) {
        e.preventDefault();
        const target = document.querySelector(this.getAttribute('href'));
        
        if (target) {
            const navHeight = document.querySelector('.professional-nav')?.offsetHeight || 0;
            const targetPosition = target.getBoundingClientRect().top + window.pageYOffset - navHeight - 20;
            
            window.scrollTo({
                top: targetPosition,
                behavior: 'smooth'
            });
        }
    });
});

// Navbar activo basado en scroll
const sections = document.querySelectorAll('section[id]');
const navLinks = document.querySelectorAll('.nav-link');

function highlightNavigation() {
    const scrollY = window.pageYOffset;
    
    sections.forEach(section => {
        const sectionHeight = section.offsetHeight;
        const sectionTop = section.offsetTop - 100;
        const sectionId = section.getAttribute('id');
        
        if (scrollY > sectionTop && scrollY <= sectionTop + sectionHeight) {
            navLinks.forEach(link => {
                link.classList.remove('active');
                if (link.getAttribute('href').includes(sectionId)) {
                    link.classList.add('active');
                }
            });
        }
    });
}

window.addEventListener('scroll', highlightNavigation);

// Animaci√≥n de entrada profesional
const observerOptions = {
    threshold: 0.1,
    rootMargin: '0px 0px -100px 0px'
};

const observer = new IntersectionObserver((entries) => {
    entries.forEach(entry => {
        if (entry.isIntersecting) {
            entry.target.style.opacity = '0';
            entry.target.style.transform = 'translateY(20px)';
            
            setTimeout(() => {
                entry.target.style.transition = 'opacity 0.6s ease, transform 0.6s ease';
                entry.target.style.opacity = '1';
                entry.target.style.transform = 'translateY(0)';
            }, 100);
            
            observer.unobserve(entry.target);
        }
    });
}, observerOptions);

// Observar elementos clave
document.addEventListener('DOMContentLoaded', () => {
    const elementsToAnimate = document.querySelectorAll('.stat-card, .service-card, .metric-card, .timeline-item');
    elementsToAnimate.forEach(el => {
        observer.observe(el);
    });
});

console.log('‚úÖ Scripts de navegaci√≥n cargados correctamente');
"""
        with open(scroll_js, 'w', encoding='utf-8') as f:
            f.write(scroll_content)
        print(f"‚úÖ Creado: assets/js/scroll.js")


def init_database():
    """Inicializa la base de datos SQLite para las consultas"""
    
    try:
        conn = sqlite3.connect(str(DB_PATH))
        cursor = conn.cursor()
        
        # Crear tabla de consultas si no existe
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS consultas (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nombre TEXT NOT NULL,
                email TEXT NOT NULL,
                empresa TEXT,
                mensaje TEXT NOT NULL,
                fecha_creacion TIMESTAMP DEFAULT CURRENT_TIMESTAMP
            )
        """)
        
        conn.commit()
        conn.close()
        print(f"‚úÖ Base de datos SQLite inicializada correctamente")
        print(f"   Ubicaci√≥n: {DB_PATH}")
        return True
    except Exception as e:
        print(f"‚ùå Error al inicializar base de datos: {e}")
        return False


def send_email_notification(nombre, email_remitente, empresa, mensaje, consulta_id):
    """Env√≠a un email de notificaci√≥n cuando se recibe una nueva consulta"""
    
    if not EMAIL_CONFIG['enable_emails']:
        print("‚ÑπÔ∏è  Notificaciones por email deshabilitadas")
        return False
    
    if not EMAIL_CONFIG['sender_password']:
        print("‚ö†Ô∏è  EMAIL_PASSWORD no configurado. No se enviar√° email.")
        print("   Para habilitar emails:")
        print("   1. Crea una App Password en Gmail")
        print("   2. Configura la variable de entorno: EMAIL_PASSWORD")
        return False
    
    try:
        # Crear el mensaje
        msg = MIMEMultipart()
        msg['From'] = EMAIL_CONFIG['sender_email']
        msg['To'] = EMAIL_CONFIG['recipient_email']
        msg['Subject'] = f"üìß Nueva Consulta Recibida - ID: {consulta_id}"
        
        # Crear el cuerpo del mensaje
        empresa_str = empresa if empresa and empresa.strip() else "No especificada"
        body = f"""
Nueva consulta recibida en tu portfolio web:

‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ
INFORMACI√ìN DE LA CONSULTA
‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ

ID de Consulta: {consulta_id}
Fecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

Nombre: {nombre}
Email: {email_remitente}
Empresa: {empresa_str}

‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ
MENSAJE
‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ

{mensaje}

‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ

Puedes responder directamente a: {email_remitente}

Esta consulta ha sido guardada autom√°ticamente en la base de datos.
"""
        
        msg.attach(MIMEText(body, 'plain', 'utf-8'))
        
        # Conectar al servidor SMTP y enviar
        server = smtplib.SMTP(EMAIL_CONFIG['smtp_server'], EMAIL_CONFIG['smtp_port'])
        server.starttls()
        server.login(EMAIL_CONFIG['sender_email'], EMAIL_CONFIG['sender_password'])
        server.send_message(msg)
        server.quit()
        
        print(f"‚úÖ Email de notificaci√≥n enviado a {EMAIL_CONFIG['recipient_email']}")
        return True
        
    except smtplib.SMTPAuthenticationError:
        print("‚ùå Error de autenticaci√≥n SMTP")
        print("   Verifica que EMAIL_PASSWORD sea una App Password v√°lida de Gmail")
        return False
    except Exception as e:
        print(f"‚ö†Ô∏è  Error al enviar email: {e}")
        print("   La consulta se guard√≥ correctamente en la base de datos")
        return False


def save_consulta(nombre, email, empresa, mensaje):
    """Guarda una consulta en la base de datos SQLite y env√≠a notificaci√≥n por email"""
    
    try:
        conn = sqlite3.connect(str(DB_PATH))
        cursor = conn.cursor()
        
        cursor.execute("""
            INSERT INTO consultas (nombre, email, empresa, mensaje)
            VALUES (?, ?, ?, ?)
        """, (nombre, email, empresa, mensaje))
        
        consulta_id = cursor.lastrowid
        conn.commit()
        conn.close()
        
        print(f"‚úÖ Consulta guardada en SQLite con ID: {consulta_id}")
        
        # Enviar notificaci√≥n por email
        send_email_notification(nombre, email, empresa, mensaje, consulta_id)
        
        return consulta_id
    except Exception as e:
        print(f"‚ùå Error al guardar en base de datos: {e}")
        return None


def kill_port_process(port):
    """Cierra autom√°ticamente el proceso que est√° usando el puerto especificado"""
    
    try:
        system = platform.system()
        
        if system == "Windows":
            # Windows: usar netstat y taskkill
            # Obtener el PID del proceso que usa el puerto
            result = subprocess.run(
                ['netstat', '-ano'],
                capture_output=True,
                text=True,
                shell=True
            )
            
            lines = result.stdout.split('\n')
            pid = None
            
            for line in lines:
                if f':{port}' in line and 'LISTENING' in line:
                    parts = line.split()
                    if len(parts) > 0:
                        pid = parts[-1]
                        break
            
            if pid:
                print(f"üîÑ Encontrado proceso PID {pid} usando el puerto {port}")
                print(f"   Cerrando proceso autom√°ticamente...")
                
                # Cerrar el proceso
                subprocess.run(
                    ['taskkill', '/PID', pid, '/F'],
                    capture_output=True,
                    shell=True
                )
                print(f"‚úÖ Proceso PID {pid} cerrado correctamente")
                return True
            else:
                print(f"‚ÑπÔ∏è  No se encontr√≥ ning√∫n proceso usando el puerto {port}")
                return False
                
        elif system == "Linux" or system == "Darwin":  # macOS
            # Linux/macOS: usar lsof y kill
            result = subprocess.run(
                ['lsof', '-ti', f':{port}'],
                capture_output=True,
                text=True
            )
            
            if result.returncode == 0 and result.stdout.strip():
                pid = result.stdout.strip()
                print(f"üîÑ Encontrado proceso PID {pid} usando el puerto {port}")
                print(f"   Cerrando proceso autom√°ticamente...")
                
                subprocess.run(['kill', '-9', pid], capture_output=True)
                print(f"‚úÖ Proceso PID {pid} cerrado correctamente")
                return True
            else:
                print(f"‚ÑπÔ∏è  No se encontr√≥ ning√∫n proceso usando el puerto {port}")
                return False
        else:
            print(f"‚ö†Ô∏è  Sistema operativo {system} no soportado para cierre autom√°tico")
            return False
            
    except Exception as e:
        print(f"‚ö†Ô∏è  Error al cerrar proceso del puerto {port}: {e}")
        return False


def validate_files():
    """Valida la existencia de archivos HTML"""
    
    required_files = ['index.html', 'cv.html', 'dashboard_marca_personal.html']
    missing = []
    
    for filename in required_files:
        filepath = Path(DIRECTORY) / filename
        if not filepath.exists():
            missing.append(filename)
    
    if missing:
        print(f"‚ö†Ô∏è  Faltaban archivos: {', '.join(missing)}")
        return False
    
    print("‚úÖ Todos los archivos HTML est√°n listos")
    return True


def check_image():
    """Verifica si existe la imagen METGO_3D.JPG"""
    
    img_path = Path(DIRECTORY) / 'METGO_3D.JPG'
    if not img_path.exists():
        print(f"\n‚ö†Ô∏è  ADVERTENCIA: No se encontr√≥ METGO_3D.JPG")
        print(f"   Por favor, coloca tu imagen profesional en: {DIRECTORY}")
        print(f"   Recomendaci√≥n: Usa una imagen profesional o logo corporativo.\n")
    else:
        print("‚úÖ Imagen METGO_3D.JPG encontrada")
```

```python
# ============================================================
# FUNCI√ìN PRINCIPAL
# ============================================================
# Variables globales para control del servidor
_httpd = None
_server_thread = None

def start_server():
    """Inicia el servidor web profesional"""
    global _httpd, _server_thread
    
    # Si ya hay un servidor corriendo, detenerlo primero
    if _httpd is not None:
        print("‚ö†Ô∏è  Ya hay un servidor corriendo. Deteni√©ndolo primero...")
        stop_server()
    
    print("=" * 70)
    print("üéØ Portfolio Profesional de Marca Personal v3.0")
    print("=" * 70)
    print("\nüìã PASO 1: Verificando archivos HTML...")
    
    # Crear archivos HTML si no existen
    create_html_files()
    
    print("\nüìã PASO 2: Creando estructura de assets...")
    
    # Crear estructura de assets
    create_assets_structure()
    

    print("\nüìã PASO 3: Inicializando base de datos SQLite...\n")

    # Inicializar base de datos SQLite
    init_database()

    print("\nüìã PASO 4: Validando archivos...\n")
    print("\nüìã PASO 4: Validando archivos...")
    
    # Validar archivos
    if not validate_files():
        print("\n‚ùå Error: Archivos HTML no pudieron crearse.")
        return None
    
    # Verificar imagen
    check_image()
    
    print("\nüìã PASO 5: Iniciando servidor web...")
    
    # Configurar servidor
    try:
        _httpd = socketserver.TCPServer(("", PORT), MetGoHandler)
        _httpd.allow_reuse_address = True  # Permitir reutilizar el puerto
        
        # Funci√≥n para ejecutar el servidor en un hilo separado
        def run_server():
            try:
                _httpd.serve_forever()
            except Exception as e:
                print(f"‚ùå Error en el servidor: {e}")
        
        # Crear y iniciar el hilo del servidor
        _server_thread = threading.Thread(target=run_server, daemon=True)
        _server_thread.start()
        
        print(f"\n{'='*70}")
        print(f"‚úÖ SERVIDOR PROFESIONAL ACTIVO")
        print(f"{'='*70}")
        print(f"\nüåê Accede a tu portfolio profesional en:")
        print(f"   üìç http://localhost:{PORT}")
        print(f"   üìç http://127.0.0.1:{PORT}")
        print(f"\nüìÑ P√°ginas disponibles:")
        print(f"   üè¢ Inicio:     http://localhost:{PORT}/index.html")
        print(f"   üìã CV:         http://localhost:{PORT}/cv.html")
        print(f"   üìä Dashboard:  http://localhost:{PORT}/dashboard_marca_personal.html")
        print(f"\nüìÇ Directorio ra√≠z: {DIRECTORY}")
        print(f"\nüí° CARACTER√çSTICAS DEL DISE√ëO PROFESIONAL:")
        print(f"   ‚úì Paleta de colores corporativa (Azul Marino)")
        print(f"   ‚úì Tipograf√≠a profesional (Inter + Playfair)")
        print(f"   ‚úì Navegaci√≥n clara entre todas las p√°ginas")
        print(f"   ‚úì Dise√±o minimalista sin iconos innecesarios")
        print(f"   ‚úì Enfoque en marca personal profesional")
        print(f"   ‚úì 100% responsive y listo para imprimir")
        print(f"\n‚èπÔ∏è  El servidor est√° corriendo en segundo plano")
        print(f"   Para detenerlo, ejecuta: stop_server()")
        print(f"{'='*70}\n")
        
        # Retornar el objeto httpd para poder cerrarlo si es necesario
        return _httpd
            
    except OSError as e:
        if e.errno == 48 or e.errno == 98 or e.errno == 10048:  # Puerto en uso (WinError 10048)
            print(f"\n{'='*70}")
            print(f"‚ö†Ô∏è  ADVERTENCIA: El puerto {PORT} ya est√° en uso")
            print(f"{'='*70}")
            print(f"\nüîÑ Cerrando proceso autom√°ticamente...")
            
            # Intentar cerrar el proceso autom√°ticamente
            if kill_port_process(PORT):
                print(f"\n‚úÖ Proceso cerrado. Reintentando iniciar servidor...")
                # Esperar un momento para que el puerto se libere
                import time
                time.sleep(1.5)
                
                # Intentar iniciar el servidor nuevamente
                try:
                    _httpd = socketserver.TCPServer(("", PORT), MetGoHandler)
                    _httpd.allow_reuse_address = True
                    
                    def run_server():
                        try:
                            _httpd.serve_forever()
                        except Exception as e:
                            print(f"‚ùå Error en el servidor: {e}")
                    
                    _server_thread = threading.Thread(target=run_server, daemon=True)
                    _server_thread.start()
                    
                    print(f"\n{'='*70}")
                    print(f"‚úÖ SERVIDOR PROFESIONAL ACTIVO")
                    print(f"{'='*70}")
                    print(f"\nüåê Accede a tu portfolio profesional en:")
                    print(f"   üìç http://localhost:{PORT}")
                    print(f"   üìç http://127.0.0.1:{PORT}")
                    print(f"\nüìÑ P√°ginas disponibles:")
                    print(f"   üè¢ Inicio:     http://localhost:{PORT}/index.html")
                    print(f"   üìã CV:         http://localhost:{PORT}/cv.html")
                    print(f"   üìä Dashboard:  http://localhost:{PORT}/dashboard_marca_personal.html")
                    print(f"\n‚èπÔ∏è  El servidor est√° corriendo en segundo plano")
                    print(f"   Para detenerlo, ejecuta: stop_server()")
                    print(f"{'='*70}\n")
                    
                    return _httpd
                except OSError as retry_error:
                    print(f"\n‚ùå Error al reintentar iniciar servidor: {retry_error}")
                    print(f"   Por favor, espera unos segundos y ejecuta start_server() nuevamente")
                    return None
            else:
                print(f"\n‚ö†Ô∏è  No se pudo cerrar el proceso autom√°ticamente.")
                print(f"   Por favor, ejecuta manualmente: stop_server()")
                print(f"   O cierra el proceso manualmente desde el Administrador de tareas")
                return None
        else:
            print(f"\n‚ùå Error al iniciar servidor: {e}")
            return None
    except Exception as e:
        print(f"\n‚ùå Error inesperado: {e}")
        return None


def stop_server():
    """Detiene el servidor web de manera controlada"""
    global _httpd, _server_thread
    
    if _httpd is None:
        print("‚ÑπÔ∏è  No hay ning√∫n servidor corriendo.")
        return
    
    try:
        print("\nüõë Deteniendo servidor...")
        
        # Cerrar el servidor de manera controlada
        if _httpd is not None:
            _httpd.shutdown()
            _httpd.server_close()
        
        # Esperar un momento para que el hilo termine
        if _server_thread is not None and _server_thread.is_alive():
            _server_thread.join(timeout=1.0)
        
        print("‚úÖ Servidor detenido correctamente")
        print("üëã ¬°Gracias por usar el Portfolio Profesional!")
    except Exception as e:
        print(f"‚ö†Ô∏è  Advertencia al detener el servidor: {e}")
    finally:
        _httpd = None
        _server_thread = None
```

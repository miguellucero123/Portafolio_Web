```python
# DASHBOARD PROFESIONAL - PROYECTO MARCA PERSONAL
# Ejecutar en Jupyter Lab

# Importar librer√≠as necesarias
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from IPython.display import HTML, display
import json
import plotly.io as pio # Importante para la plantilla

# Configuraci√≥n de estilo
colors = {
    'primary': '#1f77b4',
    'secondary': '#ff7f0e', 
    'success': '#2ca02c',
    'danger': '#d62728',
    'warning': '#ff9800',
    'info': '#17a2b8',
    'dark': '#343a40',
    'light': '#f8f9fa'
}

# --- NUEVA FUNCI√ìN: Plantilla de Estilo Global ---
def define_plotly_template():
    """
    Define y aplica una plantilla de Plotly personalizada para un look empresarial
    consistente en todos los gr√°ficos.
    """
    custom_template = go.layout.Template()
    
    # Definir la fuente principal para todo el dashboard
    main_font = 'Inter, sans-serif'
    
    custom_template.layout = go.Layout(
        # Fuentes globales
        font=dict(family=main_font, size=12, color=colors['dark']),
        
        # T√≠tulos
        title=dict(
            font=dict(family=main_font, size=22), # <--- CORREGIDO
            x=0.05, # Alinear t√≠tulos a la izquierda para un look moderno
            xanchor='left'
        ),
        
        # Fondos
        plot_bgcolor='white',
        paper_bgcolor='white',
        
        # Ejes
        xaxis=dict(
            title_font=dict(size=14), # <--- CORREGIDO
            tickfont=dict(size=12),
            gridcolor='#e0e0e0' # Rejilla sutil
        ),
        yaxis=dict(
            title_font=dict(size=14), # <--- CORREGIDO
            tickfont=dict(size=12),
            gridcolor='#e0e0e0' # Rejilla sutil
        ),
        
        # Leyenda
        legend=dict(
            font=dict(size=11),
            bgcolor='rgba(255,255,255,0.8)'
        )
    )
    
    # Registrar la plantilla
    pio.templates['personal_brand'] = custom_template
    # Establecerla como la plantilla por defecto
    pio.templates.default = 'plotly_white+personal_brand'
    
    print("üé® Plantilla de Plotly 'personal_brand' aplicada globalmente.")


# --- CSS MEJORADO: Ahora incluye un Grid System ---
def create_dashboard_css():
    """
    Crea el CSS del dashboard, ahora con un layout de CSS Grid.
    """
    return """
    <style>
        @import url('https://fonts.googleapis.com/css2?family=Inter:wght@300;400;600;700&display=swap');
        
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        
        body {
            font-family: 'Inter', sans-serif;
            background-color: #f5f7fa;
            color: #2d3748;
        }
        
        .dashboard-container {
            max-width: 1600px; /* M√°s ancho para layout empresarial */
            margin: 0 auto;
            padding: 25px;
        }
        
        .header {
            background: linear-gradient(135deg, #1f77b4 0%, #17a2b8 100%);
            color: white;
            padding: 40px;
            border-radius: 15px;
            box-shadow: 0 10px 30px rgba(0,0,0,0.1);
            margin-bottom: 30px;
            text-align: center;
        }
        
        .header h1 { font-size: 2.5rem; font-weight: 700; margin-bottom: 10px; }
        .header p { font-size: 1.2rem; opacity: 0.9; }
        
        .metrics-grid {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(250px, 1fr));
            gap: 25px;
            margin-bottom: 30px;
        }
        
        .metric-card {
            background: white;
            padding: 25px;
            border-radius: 12px;
            box-shadow: 0 4px 15px rgba(0,0,0,0.06); /* Sombra m√°s sutil */
            text-align: center;
            transition: transform 0.3s ease, box-shadow 0.3s ease;
        }
        
        .metric-card:hover {
            transform: translateY(-5px);
            box-shadow: 0 8px 25px rgba(0,0,0,0.1);
        }
        
        .metric-value { font-size: 2.5rem; font-weight: 700; margin: 10px 0; }
        .metric-label { color: #718096; font-size: 0.9rem; text-transform: uppercase; letter-spacing: 1px; }
        
        .section {
            background: white;
            padding: 30px;
            border-radius: 12px;
            box-shadow: 0 4px 15px rgba(0,0,0,0.06);
            margin-bottom: 25px;
        }
        
        .section-title {
            font-size: 1.8rem;
            font-weight: 600;
            margin-bottom: 25px;
            color: #2d3748;
            border-bottom: 3px solid #1f77b4;
            padding-bottom: 10px;
            display: inline-block;
        }

        /* --- NUEVO: CSS Grid Layout --- */
        .dashboard-grid {
            display: grid;
            grid-template-columns: repeat(3, 1fr); /* Layout de 3 columnas */
            grid-auto-rows: min-content;
            gap: 25px;
        }
        
        .grid-item {
            width: 100%;
        }
        
        .grid-col-span-1 { grid-column: span 1; }
        .grid-col-span-2 { grid-column: span 2; }
        .grid-col-span-3 { grid-column: span 3; }
        
        /* Asegura que los gr√°ficos de Plotly sean responsivos */
        .plotly-graph-div {
            width: 100% !important;
            min-height: 450px;
        }
        
        /* Clases de utilidad para el layout */
        .full-width { grid-column: 1 / -1; } /* Abarca todo el ancho */
        .main-content { grid-column: span 2; }
        .sidebar { grid-column: span 1; }
        
        /* Estilos de las secciones movidos de create_dashboard */
        .attributes-list {
            display: grid;
            grid-template-columns: 1fr; /* Una columna dentro de la sidebar */
            gap: 20px;
            margin-top: 20px;
        }
        
        .attribute-item {
            background: #f7fafc;
            padding: 20px;
            border-radius: 10px;
            border-left: 4px solid #1f77b4;
        }
        
        .attribute-number {
            background: #1f77b4;
            color: white;
            width: 35px;
            height: 35px;
            border-radius: 50%;
            display: inline-flex;
            align-items: center;
            justify-content: center;
            font-weight: 600;
            margin-right: 10px;
            flex-shrink: 0;
        }
        .attribute-item div {
            display: flex;
            align-items: center;
        }
        
        .action-grid {
            display: grid;
            grid-template-columns: 1fr 1fr;
            gap: 25px;
            margin-top: 20px;
        }
        
        .action-card {
            background: linear-gradient(135deg, #f7fafc 0%, #e9ecef 100%);
            padding: 25px;
            border-radius: 12px;
            border: 2px solid #e2e8f0;
        }
        
        .action-title { font-size: 1.3rem; font-weight: 600; color: #1f77b4; margin-bottom: 15px; }
        
        .progress-bar { width: 100%; height: 8px; background: #e2e8f0; border-radius: 4px; overflow: hidden; margin-top: 10px; }
        .progress-fill { height: 100%; background: linear-gradient(90deg, #1f77b4 0%, #17a2b8 100%); border-radius: 4px; transition: width 1s ease; }
    </style>
    """

# --- FUNCIONES DE GR√ÅFICOS (Ahora usan la plantilla global) ---

def create_timeline_chart():
    # Datos del timeline
    timeline_data = pd.DataFrame({
        'Mes': ['Mes 1', 'Mes 2', 'Mes 3', 'Mes 4'],
        'Fase': ['FUNDACI√ìN', 'ACELERACI√ìN', 'CONSOLIDACI√ìN', 'CULMINACI√ìN'],
        'Casos': ['Casos 1-2', 'Casos 3-4', 'Casos 5-6', 'Casos 7-8'],
        'Actividades': [
            'Setup perfil, Templates, Primeros an√°lisis',
            'Networking activo, Visualizaciones creativas',
            'Portfolio web, Machine Learning avanzado',
            'Casos sofisticados, Evaluaci√≥n final'
        ],
        'Color': ['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728']
    })
    
    fig = go.Figure()
    
    # A√±adir barras del timeline
    for i, row in timeline_data.iterrows():
        fig.add_trace(go.Bar(
            x=[row['Mes']],
            y=[1],
            name=row['Fase'],
            text=f"<b>{row['Fase']}</b><br>{row['Casos']}<br><i>{row['Actividades']}</i>",
            textposition='inside',
            marker_color=row['Color'],
            hovertemplate='%{text}<extra></extra>',
            showlegend=False
        ))
    
    fig.update_layout(
        title={
            'text': 'CRONOGRAMA DEL PROYECTO - 4 MESES',
            'font': {'size': 20, 'family': 'Inter, sans-serif'},
            'x': 0.5,
            'xanchor': 'center'
        },
        # --- CORRECCI√ìN AQU√ç ---
        # (Aseg√∫rate de haber corregido tambi√©n el SyntaxError anterior, 
        # usando 'xaxis=' en lugar de '"xaxis":')
        xaxis={
            "title": {"text": "Meses"},
            "showgrid": False  # <--- CORREGIDO
        },
        yaxis={
            "title": {"text": "Casos"},
            "visible": True,   # <--- CORREGIDO
            "showgrid": True   # <--- CORREGIDO
        },
        # --- FIN DE LA CORRECCI√ìN ---
        plot_bgcolor='white',
        height=400,
        margin=dict(l=0, r=0, t=60, b=40),
        barmode='group'
    )
    
    return fig

def create_metrics_projection():
    weeks = list(range(1, 17))
    engagement = [5 + (15/16)*w + np.random.normal(0, 1) for w in weeks]
    followers = [100 + (400/16)*w + np.random.normal(0, 8) for w in weeks]
    quality_connections = [10 + (90/16)*w + np.random.normal(0, 2) for w in weeks]
    
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=('Tasa de Engagement (%)', 'Crecimiento de Seguidores', 
                        'Conexiones de Calidad', 'ROI Proyectado'),
        specs=[[{'type': 'scatter'}, {'type': 'scatter'}],
               [{'type': 'scatter'}, {'type': 'bar'}]],
        vertical_spacing=0.15,
        horizontal_spacing=0.1
    )
    
    # Engagement
    fig.add_trace(go.Scatter(
        x=weeks, y=engagement, mode='lines+markers', name='Engagement',
        line=dict(color=colors['primary'], width=3),
        fill='tozeroy', fillcolor='rgba(31, 119, 180, 0.1)'
    ), row=1, col=1)
    fig.add_hline(y=20, line_dash="dash", line_color=colors['danger'], 
                  annotation_text="Meta: 20%", row=1, col=1)
    
    # Seguidores
    fig.add_trace(go.Scatter(
        x=weeks, y=followers, mode='lines+markers', name='Seguidores',
        line=dict(color=colors['success'], width=3),
        fill='tozeroy', fillcolor='rgba(44, 160, 44, 0.1)'
    ), row=1, col=2)
    
    # Conexiones de calidad
    fig.add_trace(go.Scatter(
        x=weeks, y=quality_connections, mode='lines+markers', name='Conexiones',
        line=dict(color=colors['secondary'], width=3),
        fill='tozeroy', fillcolor='rgba(255, 127, 14, 0.1)'
    ), row=2, col=1)
    
    # ROI Proyectado
    roi_data = pd.DataFrame({
        'Periodo': ['Inversi√≥n', 'A√±o 1', 'A√±o 2', 'A√±o 3'],
        'Valor': [-6612, 13600, 27200, 49800],
        'Color': [colors['danger'], colors['success'], colors['success'], colors['success']]
    })
    fig.add_trace(go.Bar(
        x=roi_data['Periodo'], y=roi_data['Valor'],
        marker_color=roi_data['Color'],
        text=[f'${v:,.0f}' for v in roi_data['Valor']],
        textposition='outside', name='ROI'
    ), row=2, col=2)
    
    fig.update_layout(
        title='PROYECCI√ìN DE M√âTRICAS CLAVE',
        showlegend=False,
        height=700,
        margin=dict(t=100)
    )
    fig.update_xaxes(title_text="Semanas", row=2, col=1)
    
    return fig

def create_content_strategy_visual():
    industries = ['Retail', 'Marketing', 'Salud', 'Social Media', 
                  'Telecom', 'Finanzas', 'Log√≠stica', 'Seguridad']
    techniques = ['An√°lisis Predictivo', 'Clustering', 'Dashboards', 
                  'NLP', 'Machine Learning', 'Series Temporales', 
                  'Optimizaci√≥n', 'Detecci√≥n Anomal√≠as']
    impact = [85, 78, 92, 75, 88, 83, 79, 90]
    
    fig = go.Figure()
    
    fig.add_trace(go.Scatterpolar(
        r=impact + [impact[0]], # Cerrar el gr√°fico
        theta=industries + [industries[0]], # Cerrar el gr√°fico
        fill='toself',
        fillcolor='rgba(31, 119, 180, 0.3)',
        line=dict(color=colors['primary'], width=2),
        text=techniques + [techniques[0]],
        hovertemplate='<b>%{theta}</b><br>T√©cnica: %{text}<br>Impacto: %{r}%<extra></extra>'
    ))
    
    fig.update_layout(
        polar=dict(
            radialaxis=dict(visible=True, range=[0, 100], ticksuffix='%')
        ),
        title='ESTRATEGIA DE CONTENIDO POR INDUSTRIA',
        height=500,
        showlegend=False,
    )
    
    return fig

def create_risk_matrix():
    risks = ['Bajo engagement inicial', 'Falta de tiempo', 'Agotamiento de ideas', 
             'Cr√≠ticas t√©cnicas', 'Competencia saturada']
    probability = [70, 60, 50, 40, 50]
    impact = [80, 80, 70, 60, 40]
    strategies = ['Promoci√≥n cruzada + Beta testing', 
                  'Batch creation + Templates',
                  'Banco 20+ ideas + Colaboraciones',
                  'Triple verificaci√≥n + Respuesta 24h',
                  'Nicho espec√≠fico + Calidad premium']
    
    fig = go.Figure()
    
    fig.add_trace(go.Scatter(
        x=probability,
        y=impact,
        mode='markers+text',
        text=risks,
        textposition='top center',
        textfont=dict(size=11, color=colors['dark']),
        marker=dict(
            size=[28, 28, 23, 18, 23],
            color=[colors['danger'], colors['danger'], colors['secondary'], colors['secondary'], colors['secondary']],
            opacity=0.8,
            line=dict(width=2, color='white')
        ),
        hovertemplate='<b>%{text}</b><br>Probabilidad: %{x}%<br>Impacto: %{y}%<br>Estrategia: %{customdata}<extra></extra>',
        customdata=strategies
    ))
    
    # Zonas de riesgo
    fig.add_shape(type="rect", x0=50, y0=50, x1=100, y1=100,
                  fillcolor="rgba(214, 39, 40, 0.15)", line=dict(width=0))
    fig.add_shape(type="rect", x0=0, y0=50, x1=50, y1=100,
                  fillcolor="rgba(255, 127, 14, 0.15)", line=dict(width=0))
    fig.add_shape(type="rect", x0=50, y0=0, x1=100, y1=50,
                  fillcolor="rgba(255, 127, 14, 0.15)", line=dict(width=0))
    fig.add_shape(type="rect", x0=0, y0=0, x1=50, y1=50,
                  fillcolor="rgba(44, 160, 44, 0.15)", line=dict(width=0))
    
    fig.update_layout(
        title="MATRIZ DE RIESGOS Y MITIGACI√ìN",
        xaxis=dict(title="Probabilidad (%)", range=[0, 100]),
        yaxis=dict(title="Impacto (%)", range=[0, 100]),
        height=500,
        showlegend=False,
    )
    
    return fig

# --- FUNCIONES AUXILIARES DE HTML (para el nuevo layout) ---

def get_header_html():
    return """
    <div class="header full-width">
        <h1>PROYECTO DE MARCA PERSONAL</h1>
        <p>T√©cnico en An√°lisis de Datos - Estrategia de Posicionamiento Profesional</p>
        <p style="margin-top: 10px; font-size: 1rem; opacity: 0.8;">Duraci√≥n: 4 meses | 8 casos de an√°lisis | Meta: 20% engagement</p>
    </div>
    """

def get_kpi_html():
    return """
    <div class="metrics-grid full-width">
        <div class="metric-card" style="border-top: 4px solid #1f77b4;">
            <div class="metric-label">Casos a Publicar</div>
            <div class="metric-value" style="color: #1f77b4;">8</div>
            <div class="progress-bar"><div class="progress-fill" style="width: 0%;"></div></div>
        </div>
        <div class="metric-card" style="border-top: 4px solid #2ca02c;">
            <div class="metric-label">Meta de Engagement</div>
            <div class="metric-value" style="color: #2ca02c;">20%</div>
            <div class="progress-bar"><div class="progress-fill" style="width: 0%;"></div></div>
        </div>
        <div class="metric-card" style="border-top: 4px solid #ff7f0e;">
            <div class="metric-label">Crecimiento de Red</div>
            <div class="metric-value" style="color: #ff7f0e;">30%</div>
            <div class="progress-bar"><div class="progress-fill" style="width: 0%;"></div></div>
        </div>
        <div class="metric-card" style="border-top: 4px solid #d62728;">
            <div class="metric-label">Duraci√≥n (meses)</div>
            <div class="metric-value" style="color: #d62728;">4</div>
            <div class="progress-bar"><div class="progress-fill" style="width: 0%;"></div></div>
        </div>
    </div>
    """

def get_executive_summary_html():
    return """
    <div style="display: grid; grid-template-columns: 1fr 2fr; gap: 30px; margin-top: 20px;">
        <div>
            <h3 style="color: #1f77b4; margin-bottom: 15px;">üéØ Objetivo Principal</h3>
            <p style="line-height: 1.6;">
                Incrementar el posicionamiento profesional como analista de datos mediante la 
                publicaci√≥n de 8 minicasos de an√°lisis en LinkedIn durante 4 meses.
            </p>
            
            <h3 style="color: #1f77b4; margin-top: 25px; margin-bottom: 15px;">üìä M√©tricas Clave</h3>
            <ul style="list-style: none; padding: 0;">
                <li style="margin-bottom: 10px;">‚Ä¢ Engagement m√≠nimo: <strong>20%</strong></li>
                <li style="margin-bottom: 10px;">‚Ä¢ Crecimiento de red: <strong>+30%</strong></li>
                <li style="margin-bottom: 10px;">‚Ä¢ Casos publicados: <strong>8</strong></li>
                <li style="margin-bottom: 10px;">‚Ä¢ Duraci√≥n: <strong>4 meses</strong></li>
            </ul>
        </div>
        
        <div>
            <h3 style="color: #1f77b4; margin-bottom: 15px;">üí° Propuesta de Valor</h3>
            <div style="background: white; padding: 20px; border-radius: 10px; margin-bottom: 20px; border: 1px solid #e2e8f0;">
                <p style="font-style: italic; font-size: 1.1rem; line-height: 1.6;">
                    "Combino pensamiento anal√≠tico avanzado con comunicaci√≥n efectiva para transformar 
                    datos complejos en insights accionables, manteniendo los m√°s altos est√°ndares √©ticos 
                    y de calidad en cada entregable."
                </p>
            </div>
        </div>
    </div>
    """

def get_attributes_html():
    return """
    <div class="attributes-list">
        <div class="attribute-item">
            <div><span class="attribute-number">1</span><strong>Pensamiento Anal√≠tico</strong></div>
            <p style="margin-top: 10px; color: #4a5568;">Capacidad para descomponer problemas complejos e identificar patrones.</p>
        </div>
        <div class="attribute-item">
            <div><span class="attribute-number">2</span><strong>Atenci√≥n Meticulosa</strong></div>
            <p style="margin-top: 10px; color: #4a5568;">Precisi√≥n excepcional en procesamiento de datos y validaci√≥n de resultados.</p>
        </div>
        <div class="attribute-item">
            <div><span class="attribute-number">3</span><strong>Comunicaci√≥n Efectiva</strong></div>
            <p style="margin-top: 10px; color: #4a5568;">Habilidad para traducir hallazgos t√©cnicos en narrativas comprensibles.</p>
        </div>
        <div class="attribute-item">
            <div><span class="attribute-number">4</span><strong>Creatividad en Visualizaci√≥n</strong></div>
            <p style="margin-top: 10px; color: #4a5568;">Dise√±o de representaciones gr√°ficas innovadoras que facilitan insights.</p>
        </div>
        <div class="attribute-item">
            <div><span class="attribute-number">5</span><strong>Compromiso √âtico</strong></div>
            <p style="margin-top: 10px; color: #4a5568;">Framework personal de √©tica de datos con principios de privacidad.</p>
        </div>
    </div>
    """

def get_action_plan_html():
    return """
    <div class="action-grid">
        <div class="action-card">
            <h3 class="action-title">üìä Acci√≥n 1: Contenido Anal√≠tico</h3>
            <ul style="list-style: none; padding-left: 0;">
                <li style="margin-bottom: 10px;">‚úì 8 casos de an√°lisis con metodolog√≠a CRISP-DM</li>
                <li style="margin-bottom: 10px;">‚úì C√≥digo reproducible en GitHub</li>
                <li style="margin-bottom: 10px;">‚úì Visualizaciones innovadoras en Tableau/Power BI</li>
                <li style="margin-bottom: 10px;">‚úì Documentaci√≥n t√©cnica exhaustiva</li>
            </ul>
        </div>
        <div class="action-card">
            <h3 class="action-title">ü§ù Acci√≥n 2: Autoridad Profesional</h3>
            <ul style="list-style: none; padding-left: 0;">
                <li style="margin-bottom: 10px;">‚úì Engagement activo < 24 horas</li>
                <li style="margin-bottom: 10px;">‚úì Participaci√≥n en 3+ grupos especializados</li>
                <li style="margin-bottom: 10px;">‚úì 8 conexiones estrat√©gicas semanales</li>
                <li style="margin-bottom: 10px;">‚úì Consistencia visual y verbal</li>
            </ul>
        </div>
    </div>
    """

def get_conclusions_html():
    return """
    <div style="background: linear-gradient(135deg, #1f77b4 0%, #17a2b8 100%); color: white; padding: 30px; border-radius: 12px;">
        <h2 class="section-title" style="color: white; border-bottom-color: white;">CONCLUSIONES Y PR√ìXIMOS PASOS</h2>
        <div style="display: grid; grid-template-columns: repeat(auto-fit, minmax(300px, 1fr)); gap: 25px; margin-top: 20px;">
            <div style="background: rgba(255,255,255,0.1); padding: 20px; border-radius: 10px;">
                <h3 style="margin-bottom: 15px;">‚úÖ Objetivos Claros</h3>
                <p>Estrategia SMART definida con m√©tricas espec√≠ficas y alcanzables en 4 meses.</p>
            </div>
            <div style="background: rgba(255,255,255,0.1); padding: 20px; border-radius: 10px;">
                <h3 style="margin-bottom: 15px;">üìà ROI Proyectado</h3>
                <p>Retorno de inversi√≥n estimado del 1,270% a 3 a√±os.</p>
            </div>
            <div style="background: rgba(255,255,255,0.1); padding: 20px; border-radius: 10px;">
                <h3 style="margin-bottom: 15px;">üöÄ Impacto Esperado</h3>
                <p>De estudiante a profesional reconocido con red de 500+ contactos relevantes.</p>
            </div>
        </div>
        <div style="text-align: center; margin-top: 30px; padding: 20px; background: rgba(255,255,255,0.2); border-radius: 10px;">
            <h3 style="font-size: 1.5rem; margin-bottom: 10px;">üéØ Visi√≥n Final</h3>
            <p style="font-size: 1.1rem; line-height: 1.6;">
                "Transformar datos en decisiones, conocimiento en valor, y conexiones en oportunidades."
            </p>
        </div>
    </div>
    """

# --- FUNCI√ìN PRINCIPAL: Construye, Exporta y Muestra el Dashboard ---

def build_and_export_dashboard(filename="dashboard_marca_personal.html"):
    """
    Construye un dashboard HTML cohesivo con un layout de grid,
    exporta el archivo y lo muestra en Jupyter.
    """
    
    # 1. Aplicar la plantilla de dise√±o global
    define_plotly_template()
    
    # 2. Generar el HTML para cada gr√°fico
    # Usamos include_plotlyjs=False porque lo cargamos una sola vez en el <head>
    # Usamos full_html=False para obtener solo el <div> del gr√°fico
    timeline_html = pio.to_html(create_timeline_chart(), include_plotlyjs=False, full_html=False)
    metrics_html = pio.to_html(create_metrics_projection(), include_plotlyjs=False, full_html=False)
    content_html = pio.to_html(create_content_strategy_visual(), include_plotlyjs=False, full_html=False)
    risk_html = pio.to_html(create_risk_matrix(), include_plotlyjs=False, full_html=False)
    
    # 3. Ensamblar el documento HTML final
    final_html = f"""
    <html>
    <head>
        <meta charset="utf-8">
        <title>Dashboard Marca Personal</title>
        <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
        {create_dashboard_css()}
    </head>
    <body>
        <div class="dashboard-container">
            
            {get_header_html()}
            
            {get_kpi_html()}
            
            <div class="dashboard-grid">
            
                <div class="grid-item main-content">
                    <div class="section" style="background: #f7fafc; border: 1px solid #e2e8f0;">
                        <h2 class="section-title">RESUMEN EJECUTIVO</h2>
                        {get_executive_summary_html()}
                    </div>
                    <div class="section">
                        <h2 class="section-title">CRONOGRAMA DEL PROYECTO</h2>
                        {timeline_html}
                    </div>
                    <div class="section">
                        <h2 class="section-title">PROYECCI√ìN DE M√âTRICAS CLAVE</h2>
                        {metrics_html}
                    </div>
                </div>
                
                <div class="grid-item sidebar">
                    <div class="section">
                        <h2 class="section-title">MIS 5 ATRIBUTOS</h2>
                        {get_attributes_html()}
                    </div>
                    <div class="section">
                        <h2 class="section-title">ESTRATEGIA DE CONTENIDO</h2>
                        {content_html}
                    </div>
                    <div class="section">
                        <h2 class="section-title">MATRIZ DE RIESGOS</h2>
                        {risk_html}
                    </div>
                </div>
                
                <div class="grid-item full-width">
                    <div class="section">
                        <h2 class="section-title">PLAN DE ACCI√ìN ESTRAT√âGICO</h2>
                        {get_action_plan_html()}
                    </div>
                </div>
                
                <div class="grid-item full-width">
                    {get_conclusions_html()}
                </div>
            
            </div> <footer style="text-align:center; margin-top:30px; padding: 20px; color:#718096;">
                <p>üìä Dashboard de Marca Personal | Creado con Python, Plotly & CSS Grid</p>
            </footer>
            
        </div> </body>
    </html>
    """
    
    # 4. Escribir el archivo
    try:
        with open(filename, "w", encoding="utf-8") as f:
            f.write(final_html)
        print(f"‚úÖ Dashboard exportado exitosamente como '{filename}'")
    except Exception as e:
        print(f"‚ùå Error al exportar el archivo: {e}")
        
    # 5. Mostrar el dashboard en Jupyter
    # Usamos display() para renderizar el HTML directamente
    display(HTML(final_html))

# --- EJECUCI√ìN DEL DASHBOARD ---
# Llama a esta funci√≥n para construir, exportar y mostrar tu dashboard
build_and_export_dashboard()
```

```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
import plotly.io as pio

def export_dashboard_pptx(filename="dashboard.pptx"):
    # Exportar gr√°ficos como im√°genes
    timeline_fig = create_timeline_chart()
    metrics_fig = create_metrics_projection()
    content_fig = create_content_strategy_visual()
    risk_fig = create_risk_matrix()

    timeline_fig.write_image("timeline.png", width=1200, height=700)
    metrics_fig.write_image("metrics.png", width=1200, height=700)
    content_fig.write_image("content.png", width=1200, height=700)
    risk_fig.write_image("risk.png", width=1200, height=700)

    # Crear presentaci√≥n
    prs = Presentation()

    # --- Diapositiva de t√≠tulo ---
    slide_layout = prs.slide_layouts[0]  # Title Slide
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    subtitle = slide.placeholders[1]

    title.text = "üìä Dashboard de Marca Personal"
    subtitle.text = "Reporte generado con Python + Plotly"

    # --- Resumen Ejecutivo ---
    slide_layout = prs.slide_layouts[1]  # Title + Content
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Resumen Ejecutivo"
    content.text = (
        "Este dashboard integra estrategia de marca personal con visualizaciones "
        "para an√°lisis de timeline, m√©tricas proyectadas, estrategia de contenido "
        "y matriz de riesgos."
    )

    # --- Timeline ---
    slide_layout = prs.slide_layouts[5]  # Blank
    slide = prs.slides.add_slide(slide_layout)
    left, top, width = Inches(1), Inches(1), Inches(8)
    slide.shapes.add_picture("timeline.png", left, top, width=width)

    # --- M√©tricas ---
    slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.add_picture("metrics.png", left, top, width=width)

    # --- Estrategia de Contenido ---
    slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.add_picture("content.png", left, top, width=width)

    # --- Matriz de Riesgos ---
    slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.add_picture("risk.png", left, top, width=width)

    # --- Conclusiones ---
    slide_layout = prs.slide_layouts[1]  # Title + Content
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Conclusiones"
    content.text = (
        "‚úÖ Estrategia clara y progresiva en el tiempo\n"
        "‚úÖ M√©tricas cuantitativas para medir impacto\n"
        "‚úÖ Estrategia de contenido alineada a objetivos\n"
        "‚úÖ Identificaci√≥n y mitigaci√≥n de riesgos"
    )

    # Guardar presentaci√≥n
    prs.save(filename)
    print(f"‚úÖ Dashboard exportado como {filename}")
```

```python
import plotly.graph_objects as go

def create_timeline_chart():
    fig = go.Figure()
    fig.add_trace(go.Bar(x=["Fase 1","Fase 2"], y=[10,20]))
    return fig

def create_metrics_projection():
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=[1,2,3], y=[2,5,9]))
    return fig

def create_content_strategy_visual():
    fig = go.Figure()
    fig.add_trace(go.Pie(labels=["Blog","Video","Podcast"], values=[40,35,25]))
    return fig

def create_risk_matrix():
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=[1,2,3], y=[3,2,1], mode="markers"))
    return fig
```

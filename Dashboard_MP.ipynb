```python
# --- PASO 1: GENERADOR DEL DASHBOARD CON BOTÓN AL CV ---
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np
from IPython.display import HTML, display
import plotly.io as pio

# 1. Configuración y Estilos Globales
colors = {'primary': '#1f77b4', 'secondary': '#ff7f0e', 'success': '#2ca02c', 
          'danger': '#d62728', 'dark': '#343a40', 'light': '#f8f9fa'}

def define_plotly_template():
    custom_template = go.layout.Template()
    custom_template.layout = go.Layout(
        font=dict(family='Inter, sans-serif', size=12, color=colors['dark']),
        title=dict(font=dict(size=22), x=0.05, xanchor='left'),
        plot_bgcolor='white', paper_bgcolor='white',
        xaxis=dict(gridcolor='#e0e0e0'), yaxis=dict(gridcolor='#e0e0e0')
    )
    pio.templates['personal_brand'] = custom_template
    pio.templates.default = 'plotly_white+personal_brand'

def create_dashboard_css():
    return """
    <style>
        @import url('https://fonts.googleapis.com/css2?family=Inter:wght@300;400;600;700&display=swap');
        * { margin: 0; padding: 0; box-sizing: border-box; }
        body { font-family: 'Inter', sans-serif; background-color: #f5f7fa; color: #2d3748; }
        .dashboard-container { max-width: 1600px; margin: 0 auto; padding: 25px; }
        
        /* HEADER Y BOTÓN */
        .header {
            background: linear-gradient(135deg, #1f77b4 0%, #17a2b8 100%);
            color: white; padding: 40px; border-radius: 15px;
            box-shadow: 0 10px 30px rgba(0,0,0,0.1); margin-bottom: 30px;
            text-align: center;
        }
        .cv-button {
            display: inline-block; margin-top: 25px; padding: 12px 30px;
            background-color: white; color: #1f77b4; font-weight: 700;
            text-decoration: none; border-radius: 50px;
            box-shadow: 0 4px 15px rgba(0,0,0,0.2); transition: all 0.3s ease;
        }
        .cv-button:hover { transform: translateY(-3px); box-shadow: 0 8px 20px rgba(0,0,0,0.3); }

        /* LAYOUT */
        .dashboard-grid { display: grid; grid-template-columns: repeat(3, 1fr); gap: 25px; }
        .grid-item { width: 100%; }
        .full-width { grid-column: 1 / -1; }
        .main-content { grid-column: span 2; }
        .sidebar { grid-column: span 1; }
        
        /* COMPONENTES */
        .section { background: white; padding: 30px; border-radius: 12px; margin-bottom: 25px; box-shadow: 0 4px 15px rgba(0,0,0,0.06); }
        .section-title { font-size: 1.8rem; font-weight: 600; margin-bottom: 25px; border-bottom: 3px solid #1f77b4; display: inline-block; }
        .metric-card { background: white; padding: 25px; border-radius: 12px; text-align: center; box-shadow: 0 4px 15px rgba(0,0,0,0.06); }
        .metrics-grid { display: grid; grid-template-columns: repeat(auto-fit, minmax(200px, 1fr)); gap: 25px; margin-bottom: 30px; }
        .metric-value { font-size: 2.5rem; font-weight: 700; margin: 10px 0; }
        
        /* UTILIDADES */
        .plotly-graph-div { width: 100% !important; min-height: 450px; }
    </style>
    """

# 2. Funciones de Gráficos
def create_all_charts():
    # Timeline
    timeline = pd.DataFrame({
        'Mes': ['Mes 1', 'Mes 2', 'Mes 3', 'Mes 4'],
        'Fase': ['FUNDACIÓN', 'ACELERACIÓN', 'CONSOLIDACIÓN', 'CULMINACIÓN'],
        'Color': ['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728']
    })
    fig_time = go.Figure()
    for _, row in timeline.iterrows():
        fig_time.add_trace(go.Bar(x=[row['Mes']], y=[1], name=row['Fase'], marker_color=row['Color']))
    fig_time.update_layout(title='CRONOGRAMA', xaxis={'visible':True}, yaxis={'visible':False}, height=400, showlegend=True)

    # Métricas
    fig_metrics = make_subplots(rows=2, cols=2, subplot_titles=('Engagement', 'Seguidores', 'Conexiones', 'ROI'))
    weeks = list(range(1,17))
    fig_metrics.add_trace(go.Scatter(x=weeks, y=[5+w for w in weeks], name='Eng', fill='tozeroy'), row=1, col=1)
    fig_metrics.add_trace(go.Scatter(x=weeks, y=[100+20*w for w in weeks], name='Seg', line=dict(color='green')), row=1, col=2)
    fig_metrics.add_trace(go.Scatter(x=weeks, y=[10+5*w for w in weeks], name='Con', line=dict(color='orange')), row=2, col=1)
    fig_metrics.add_trace(go.Bar(x=['Inv', 'A1', 'A2'], y=[-6000, 13000, 27000], name='ROI'), row=2, col=2)
    fig_metrics.update_layout(height=600, showlegend=False, title='PROYECCIÓN')

    # Estrategia (Radar)
    fig_radar = go.Figure(go.Scatterpolar(
        r=[80, 70, 90, 60, 80], theta=['Retail', 'Salud', 'Finanzas', 'Logística', 'Retail'], fill='toself'
    ))
    fig_radar.update_layout(polar=dict(radialaxis=dict(visible=True, range=[0, 100])), title='ESTRATEGIA', height=400)

    # Riesgos
    fig_risk = go.Figure(go.Scatter(
        x=[70, 60, 50], y=[80, 80, 70], mode='markers+text', text=['Engagement', 'Tiempo', 'Ideas'],
        marker=dict(size=30, color=['red', 'red', 'orange'])
    ))
    fig_risk.update_layout(title="MATRIZ RIESGOS", xaxis=dict(range=[0,100], title='Probabilidad'), yaxis=dict(range=[0,100], title='Impacto'), height=400)

    return fig_time, fig_metrics, fig_radar, fig_risk

# 3. Construcción del HTML
def build_dashboard():
    define_plotly_template()
    f1, f2, f3, f4 = create_all_charts()
    
    html_content = f"""
    <html>
    <head><meta charset="utf-8"><title>Miguel Lucero Dashboard</title>{create_dashboard_css()}<script src="https://cdn.plot.ly/plotly-latest.min.js"></script></head>
    <body>
        <div class="dashboard-container">
            <div class="header">
                <h1>PROYECTO MARCA PERSONAL</h1>
                <p>Estrategia de Posicionamiento - Data Analyst</p>
                <a href="miguel_lucero_cv.html" target="_blank" class="cv-button">Ver CV Profesional &rarr;</a>
            </div>
            
            {get_kpi_html()} <div class="dashboard-grid">
                <div class="grid-item main-content">
                    <div class="section">{pio.to_html(f1, full_html=False, include_plotlyjs=False)}</div>
                    <div class="section">{pio.to_html(f2, full_html=False, include_plotlyjs=False)}</div>
                </div>
                <div class="grid-item sidebar">
                    <div class="section">{pio.to_html(f3, full_html=False, include_plotlyjs=False)}</div>
                    <div class="section">{pio.to_html(f4, full_html=False, include_plotlyjs=False)}</div>
                </div>
            </div>
        </div>
    </body>
    </html>
    """
    
    with open("dashboard_marca_personal.html", "w", encoding="utf-8") as f:
        f.write(html_content)
    print("✅ Dashboard creado: abre 'dashboard_marca_personal.html'")
    display(HTML(html_content))

def get_kpi_html():
    return """
    <div class="metrics-grid">
        <div class="metric-card" style="border-top: 4px solid #1f77b4;"><div class="metric-value">8</div><div>Casos</div></div>
        <div class="metric-card" style="border-top: 4px solid #2ca02c;"><div class="metric-value">20%</div><div>Engagement</div></div>
        <div class="metric-card" style="border-top: 4px solid #ff7f0e;"><div class="metric-value">30%</div><div>Red</div></div>
    </div>"""

# EJECUTAR
build_dashboard()
```

```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
import plotly.io as pio

def export_dashboard_pptx(filename="dashboard.pptx"):
    # Exportar gráficos como imágenes
    timeline_fig = create_timeline_chart()
    metrics_fig = create_metrics_projection()
    content_fig = create_content_strategy_visual()
    risk_fig = create_risk_matrix()

    timeline_fig.write_image("timeline.png", width=1200, height=700)
    metrics_fig.write_image("metrics.png", width=1200, height=700)
    content_fig.write_image("content.png", width=1200, height=700)
    risk_fig.write_image("risk.png", width=1200, height=700)

    # Crear presentación
    prs = Presentation()

    # --- Diapositiva de título ---
    slide_layout = prs.slide_layouts[0]  # Title Slide
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    subtitle = slide.placeholders[1]

    title.text = "📊 Dashboard de Marca Personal"
    subtitle.text = "Reporte generado con Python + Plotly"

    # --- Resumen Ejecutivo ---
    slide_layout = prs.slide_layouts[1]  # Title + Content
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Resumen Ejecutivo"
    content.text = (
        "Este dashboard integra estrategia de marca personal con visualizaciones "
        "para análisis de timeline, métricas proyectadas, estrategia de contenido "
        "y matriz de riesgos."
    )

    # --- Timeline ---
    slide_layout = prs.slide_layouts[5]  # Blank
    slide = prs.slides.add_slide(slide_layout)
    left, top, width = Inches(1), Inches(1), Inches(8)
    slide.shapes.add_picture("timeline.png", left, top, width=width)

    # --- Métricas ---
    slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.add_picture("metrics.png", left, top, width=width)

    # --- Estrategia de Contenido ---
    slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.add_picture("content.png", left, top, width=width)

    # --- Matriz de Riesgos ---
    slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.add_picture("risk.png", left, top, width=width)

    # --- Conclusiones ---
    slide_layout = prs.slide_layouts[1]  # Title + Content
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Conclusiones"
    content.text = (
        "✅ Estrategia clara y progresiva en el tiempo\n"
        "✅ Métricas cuantitativas para medir impacto\n"
        "✅ Estrategia de contenido alineada a objetivos\n"
        "✅ Identificación y mitigación de riesgos"
    )

    # Guardar presentación
    prs.save(filename)
    print(f"✅ Dashboard exportado como {filename}")
```

```python
import plotly.graph_objects as go

def create_timeline_chart():
    fig = go.Figure()
    fig.add_trace(go.Bar(x=["Fase 1","Fase 2"], y=[10,20]))
    return fig

def create_metrics_projection():
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=[1,2,3], y=[2,5,9]))
    return fig

def create_content_strategy_visual():
    fig = go.Figure()
    fig.add_trace(go.Pie(labels=["Blog","Video","Podcast"], values=[40,35,25]))
    return fig

def create_risk_matrix():
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=[1,2,3], y=[3,2,1], mode="markers"))
    return fig
```
